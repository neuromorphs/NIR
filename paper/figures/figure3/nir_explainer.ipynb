```python
spk_t = [10, 13, 19, 32, 34, 37, 46, 48, 51, 55]
v_t = [0]
decay = 0.9
thr = 2.5
spk = []
for t in range(65):
    v = v_t[-1] * decay
    if t in spk_t:
        v += 1
    v_t.append(v)
    if v_t[-1] >= thr:
        spk.append(t)
        v_t[-1] = 0

plt.figure(figsize=(10, 3))
fig, ax = plt.subplots(figsize=(10, 3))
ax.eventplot(spk, color='k', linelengths=0.3, linewidths=4)
ax.grid(0, which='both')
ax.set_xlim(0, 65)
ax.set_ylim(0.5, 1.5)

# Hide the right and top spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# # Hide the ticks
# ax.set_xticks([])
# ax.set_yticks([])
# Hide the labels
ax.set_xticklabels([])
ax.set_yticklabels([])
# Draw arrows
ax.plot(1, 0.5, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

fig.savefig('lif_output_ev.pdf', bbox_inches='tight', dpi=300)

spk_t = [10, 13, 19, 32, 34, 37, 46, 48, 51, 55]
v_t = [0]
decay = 0.9
thr = 2.5
spk = [0]
for t in range(65):
    v = v_t[-1] * decay
    if t in spk_t:
        v += 1
    v_t.append(v)
    if v_t[-1] >= thr:
        spk.append(1)
        v_t[-1] = 0
    else:
        spk.append(0)


plt.figure(figsize=(10, 3))
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(spk, color='k', linewidth=4)
ax.grid(0, which='both')
ax.set_xlim(0, 65)
ax.set_ylim(0.0, 1.5)

# Hide the right and top spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# # Hide the ticks
# ax.set_xticks([])
# ax.set_yticks([])
# Hide the labels
ax.set_xticklabels([])
ax.set_yticklabels([])
# Draw arrows
ax.plot(1, 0.5, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

fig.savefig('lif_output_trace.pdf', bbox_inches='tight', dpi=300)


plt.figure(figsize=(10, 3))
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot([-e for e in spk], color='k', linewidth=4)
ax.grid(0, which='both')
ax.set_xlim(0, 65)
ax.set_ylim(-1.5, 0.0)

# Hide the right and top spines
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_color('none')
# # Hide the ticks
# ax.set_xticks([])
# ax.set_yticks([])
# Hide the labels
ax.set_xticklabels([])
ax.set_yticklabels([])
# Draw arrows
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "vk", transform=ax.get_xaxis_transform(), clip_on=False)

fig.savefig('lif_output_trace_inv.pdf', bbox_inches='tight', dpi=300)
```

```python
v_t = [0]
decay = 0.9
for t in range(65):
    v = v_t[-1] * decay
    if t in spk_t:
        v += 1
    v_t.append(v)

for t in range(65):
    # low pass filter the v_t signal
    v_t[t] = np.mean(v_t[max(0, t-1):t+1])

fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(v_t, color='k', linewidth=4)
ax.set_ylim(0, 4)
ax.set_xlim(0, 65)

# Hide the right and top spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# # Hide the ticks
# ax.set_xticks([])
# ax.set_yticks([])
# Hide the labels
ax.set_xticklabels([])
ax.set_yticklabels([])
# Draw arrows
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
fig.savefig('linear_in.pdf', dpi=300, bbox_inches='tight')
plt.close()


# now plot the same thing, but in negative
v_t = [-e for e in v_t]

fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(v_t, color='k', linewidth=4)
ax.set_ylim(-4, 0)
ax.set_xlim(0, 65)

# Hide the right and top spines
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_color('none')
# Hide the labels
ax.set_xticklabels([])
ax.set_yticklabels([])
# Draw arrows
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 0, "vk", transform=ax.get_xaxis_transform(), clip_on=False)
fig.savefig('linear_out.pdf', dpi=300, bbox_inches='tight')
```
