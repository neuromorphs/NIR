```python
spk_t = [10, 13, 19, 32, 34, 37, 46, 48, 51, 55]
v_t = [0]
decay = 0.9
thr = 2.5
spk = []
for t in range(65):
    v = v_t[-1] * decay
    if t in spk_t:
        v += 1
    v_t.append(v)
    if v_t[-1] >= thr:
        spk.append(t)
        v_t[-1] = 0

fig, ax = plt.subplots(figsize=(10, 3))
ax.eventplot(spk, color='k', linelengths=0.3, linewidths=EV_LW)
ax.grid(0, which='both')
ax.set_xlim(0, 65)
ax.set_ylim(0.5, 1.5)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_linewidth(SP_LW)
ax.spines['left'].set_linewidth(SP_LW)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xticklabels([])
ax.set_yticklabels([])

ax.plot(1, 0.5, ">k", transform=ax.get_yaxis_transform(), clip_on=False, markersize=ARR_S)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False, markersize=ARR_S)

fig.savefig('lif_output_ev.pdf', bbox_inches='tight', dpi=300)
```

```python
v_t = [0]
decay = 0.9
for t in range(65):
    v = v_t[-1] * decay
    if t in spk_t:
        v += 1
    v_t.append(v)

for t in range(65):
    # low pass filter the v_t signal
    v_t[t] = np.mean(v_t[max(0, t-1):t+1])

fig, ax = plt.subplots(figsize=(10, 3))
xt = range(X_OFF, len(v_t)+X_OFF)
yt = [v+Y_OFF*1.5 for v in v_t]
ax.plot(xt, yt, color='k', linewidth=EV_LW)
ax.set_ylim(0, 4)
ax.set_xlim(0, 65)

# Hide the right and top spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_linewidth(SP_LW)
ax.spines['bottom'].set_linewidth(SP_LW)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xticklabels([])
ax.set_yticklabels([])
# Draw arrows
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False, markersize=ARR_S)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False, markersize=ARR_S)
fig.savefig('linear_in.pdf', dpi=300, bbox_inches='tight')
```

```python
dt = 1e-3
r, sigma, b = 28, 12, 8/3
T = 20000
tt = np.arange(T)
xt = np.ones((T))
yt = np.ones((T))
zt = np.ones((T)) * 20
for t in tt[:-1]:
    xt[t+1] = xt[t] + (sigma * (yt[t]-xt[t])) * dt
    yt[t+1] = yt[t] + (r*xt[t] - yt[t] - xt[t]*zt[t]) * dt
    zt[t+1] = zt[t] + (xt[t]*yt[t] - b*zt[t]) * dt
# plt.plot(tt, xt)
# plt.plot(tt, yt)
# plt.plot(tt, zt)
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(xt, zt, color='k', linewidth=EV_LW/4)
ax.set_ylim(6, 45)
ax.set_xlim(-20, 20)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_linewidth(SP_LW)
ax.spines['bottom'].set_linewidth(SP_LW)
ax.set_xticks([])
ax.set_yticks([])
ax.plot(20, 6, ">k", clip_on=False, markersize=ARR_S)
ax.plot(-20, 45, "^k", clip_on=False, markersize=ARR_S)
fig.savefig('hidden.pdf', dpi=300, bbox_inches='tight')
```
